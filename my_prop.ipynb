```python
import cv2
import os

def extract_rois(image_path, boxes, output_dir, label):
    """Extract ROIs from YOLO bounding boxes and save them as labeled images."""
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Cannot read image at {image_path}")
        return

    image_name = os.path.basename(image_path).split('.')[0]

    # Ensure the output directory exists
    save_dir = os.path.join(output_dir, label)
    os.makedirs(save_dir, exist_ok=True)

    for i, box in enumerate(boxes):
        x_min, y_min, x_max, y_max = map(int, box)
        roi = image[y_min:y_max, x_min:x_max]

        if roi.size == 0:  # Skip empty or invalid ROIs
            print(f"Warning: Empty ROI for box {box} in image {image_path}")
            continue

        roi_path = os.path.join(save_dir, f"{image_name}_roi{i}.jpg")
        cv2.imwrite(roi_path, roi)

def process_image_for_vit(image_path, annotation_path, output_dir, label="benign"):
    """Process a single image and annotation file to extract ROIs."""
    # Read YOLO annotation file
    with open(annotation_path, 'r') as file:
        lines = file.readlines()

    # Load the image to get its dimensions
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Cannot read image at {image_path}")
        return
    image_height, image_width = image.shape[:2]

    boxes = []
    for line in lines:
        parts = line.strip().split()
        class_id, x_center, y_center, width, height = map(float, parts)

        # Convert YOLO normalized coordinates to pixel values
        x_center *= image_width
        y_center *= image_height
        width *= image_width
        height *= image_height

        x_min = x_center - (width / 2)
        y_min = y_center - (height / 2)
        x_max = x_center + (width / 2)
        y_max = y_center + (height / 2)

        boxes.append([x_min, y_min, x_max, y_max])

    # Extract ROIs from the full image based on bounding boxes
    extract_rois(image_path, boxes, output_dir, label)

# Example usage
image_path = "/content/dataset/train/image/12.JPG"
annotation_path = "/content/dataset/train/label/12.txt"
output_dir = "/content/dataset/vit/train"

# Specify the label dynamically (e.g., "benign" or "malignant")
process_image_for_vit(image_path, annotation_path, output_dir, label="benign")
```

```python
import torch
from PIL import Image
import cv2
import numpy as np
from transformers import ViTForImageClassification, ViTFeatureExtractor

# Step 1: YOLO Model Setup (Using YOLOv8 from Ultralytics)
from ultralytics import YOLO

# Load YOLOv8 model (pre-trained or fine-tuned on your dataset)
yolo_model = YOLO("/content/yolov8n.pt")  # Replace with your model path

# Step 2: ViT Model Setup
# Load pre-trained ViT model and feature extractor
vit_model = ViTForImageClassification.from_pretrained("google/vit-base-patch16-224")
vit_feature_extractor = ViTFeatureExtractor.from_pretrained("google/vit-base-patch16-224")

# Step 3: Load and Process the Input Image
def preprocess_image(image_path):
    """Load and preprocess the input image for YOLO."""
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB
    return image, image_rgb

# Step 4: Detect Tumors using YOLO
def detect_tumors(image):
    """Run YOLO detection and return bounding boxes."""
    results = yolo_model(image)
    boxes = results[0].boxes.xyxy.cpu().numpy()  # Extract bounding boxes
    scores = results[0].boxes.conf.cpu().numpy()  # Extract confidence scores
    return boxes, scores

# Step 5: Extract and Resize ROIs for ViT
def extract_and_resize_rois(image, boxes, input_size=224):
    """Extract ROIs and resize for ViT."""
    rois = []
    for box in boxes:
        x_min, y_min, x_max, y_max = map(int, box)
        roi = image[y_min:y_max, x_min:x_max]  # Crop ROI
        roi_resized = cv2.resize(roi, (input_size, input_size))  # Resize to ViT input size
        rois.append(roi_resized)
    return rois

# Step 6: Classify ROIs using ViT
def classify_rois(rois):
    """Classify each ROI using ViT."""
    results = []
    for roi in rois:
        # Convert ROI to PIL Image and normalize using ViT feature extractor
        roi_pil = Image.fromarray(roi)
        inputs = vit_feature_extractor(images=roi_pil, return_tensors="pt")
        outputs = vit_model(**inputs)
        predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
        predicted_class = torch.argmax(predictions, dim=-1).item()
        confidence = predictions[0][predicted_class].item()
        results.append((predicted_class, confidence))
    return results

# Step 7: Full Pipeline
def tumor_detection_and_classification(image_path):
    """Full pipeline: Detect tumors and classify them."""
    # Preprocess and detect tumors
    original_image, processed_image = preprocess_image(image_path)
    boxes, scores = detect_tumors(processed_image)

    # Extract and classify ROIs
    rois = extract_and_resize_rois(original_image, boxes)
    classifications = classify_rois(rois)

    # Display results
    for i, (box, score, classification) in enumerate(zip(boxes, scores, classifications)):
        x_min, y_min, x_max, y_max = map(int, box)
        label, confidence = classification
        print(f"Tumor {i + 1}:")
        print(f" - Bounding Box: [{x_min}, {y_min}, {x_max}, {y_max}]")
        print(f" - YOLO Confidence: {score:.2f}")
        print(f" - Classification: {label} (Confidence: {confidence:.2f})")

# Run the pipeline on an example image
image_path = "/content/my-project-4/test/102_PNG.rf.c169f719d2781aaaf9101a5110b74b56.jpg"  # Replace with your image path
tumor_detection_and_classification(image_path)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, confusion_matrix
import numpy as np

# Example: Ground truth and predictions
ground_truth = [
    {"bbox": [50, 50, 200, 200], "label": "Benign"},  # x_min, y_min, x_max, y_max
    {"bbox": [300, 300, 400, 400], "label": "Malignant"}
]

predictions = [
    {"bbox": [55, 55, 195, 195], "label": "Benign", "confidence": 0.85},
    {"bbox": [310, 310, 390, 390], "label": "Benign", "confidence": 0.56}
]

# Compute IoU for bounding boxes
def compute_iou(box1, box2):
    x_min1, y_min1, x_max1, y_max1 = box1
    x_min2, y_min2, x_max2, y_max2 = box2

    # Compute overlap
    xi1 = max(x_min1, x_min2)
    yi1 = max(y_min1, y_min2)
    xi2 = min(x_max1, x_max2)
    yi2 = min(y_max1, y_max2)
    inter_area = max(0, xi2 - xi1) * max(0, yi2 - yi1)

    # Compute union
    box1_area = (x_max1 - x_min1) * (y_max1 - y_min1)
    box2_area = (x_max2 - x_min2) * (y_max2 - y_min2)
    union_area = box1_area + box2_area - inter_area

    return inter_area / union_area if union_area > 0 else 0

# Match predictions with ground truth
y_true = []
y_pred = []
ious = []

for gt in ground_truth:
    gt_bbox = gt["bbox"]
    gt_label = gt["label"]

    matched = False
    for pred in predictions:
        iou = compute_iou(gt_bbox, pred["bbox"])
        if iou > 0.5:  # Match if IoU > 0.5
            ious.append(iou)
            y_true.append(gt_label)
            y_pred.append(pred["label"])
            matched = True
            break

    if not matched:  # False negative
        y_true.append(gt_label)
        y_pred.append("None")  # No prediction

# Handle false positives
for pred in predictions:
    if all(compute_iou(pred["bbox"], gt["bbox"]) <= 0.5 for gt in ground_truth):
        y_true.append("None")
        y_pred.append(pred["label"])

# Compute metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average="weighted", zero_division=0)
recall = recall_score(y_true, y_pred, average="weighted", zero_division=0)
f1 = f1_score(y_true, y_pred, average="weighted", zero_division=0)
iou = np.mean(ious)
# Convert to percentage and format
accuracy_pct = f"{accuracy * 100:.2f}%"
precision_pct = f"{precision * 100:.2f}%"
recall_pct = f"{recall * 100:.2f}%"
f1_pct = f"{f1 * 100:.2f}%"
iou_pct = f"{iou * 100:.2f}%"

# Print metrics in percentage
print(f"Accuracy: {accuracy_pct}")
print(f"Precision: {precision_pct}")
print(f"Recall: {recall_pct}")
print(f"F1-Score: {f1_pct}")
print(f"IoU: {iou_pct}")
```

```python
import cv2
import matplotlib.pyplot as plt

def detect_tumors(image_path, yolo_model):
    """
    Detect tumors in an image using a YOLO model.

    Args:
        image_path (str): Path to the image.
        yolo_model: YOLO model instance.

    Returns:
        tuple: Processed image and list of (ROI, bbox) tuples.
    """
    # Load the image
    image = cv2.imread(image_path)

    # Perform YOLO inference
    results = yolo_model.predict(image)
    boxes = results.boxes.xyxy.cpu().numpy()  # Extract bounding boxes

    # Extract ROIs and their bounding boxes
    rois = []
    for box in boxes:
        x_min, y_min, x_max, y_max = map(int, box[:4])  # Get bbox coordinates
        roi = image[y_min:y_max, x_min:x_max]  # Crop ROI
        rois.append((roi, (x_min, y_min, x_max, y_max)))  # Append (ROI, bbox)

    return image, rois

def classify_tumor(roi, vit_model, vit_processor):
    """
    Classify a tumor using the ViT model.

    Args:
        roi: Region of interest (cropped image).
        vit_model: ViT model instance.
        vit_processor: ViT image processor.

    Returns:
        tuple: Predicted label and confidence score.
    """
    # Preprocess the ROI
    inputs = vit_processor(images=roi, return_tensors="pt")

    # Perform classification
    outputs = vit_model(**inputs)
    logits = outputs.logits
    confidence = logits.softmax(dim=1).max().item()
    label = logits.argmax(dim=1).item()

    # Map label index to class name
    class_names = ["Benign", "Malignant"]
    label_name = class_names[label]

    return label_name, confidence

def hybrid_pipeline(image_path, yolo_model, vit_model, vit_processor):
    """
    Hybrid pipeline for tumor detection and classification.
    """
    # Step 1: Tumor Detection with YOLO
    image, rois = detect_tumors(image_path, yolo_model)

    # Step 2: Classification of Detected Tumors with ViT
    for roi, bbox in rois:
        label, confidence = classify_tumor(roi, vit_model, vit_processor)
        print(f"Tumor detected: {label} (Confidence: {confidence:.2f})")

        # Draw bounding box and classification label
        x_min, y_min, x_max, y_max = bbox
        cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)
        cv2.putText(image, f"{label} ({confidence:.2f})", (x_min, y_min - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Show the final result
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for plt
    plt.axis('off')
    plt.show()

# Test the pipeline
test_image_path = "/content/dataset/train/image/12.JPG"
hybrid_pipeline(test_image_path, yolo_model, vit_model, vit_processor)
```

```python
import cv2
import os

def extract_rois(image_path, boxes, output_dir, label):
    """Extract ROIs from YOLO bounding boxes and save them as labeled."""
    image = cv2.imread(image_path)
    image_name = os.path.basename(image_path).split('.')[0]
    image_height, image_width, _ = image.shape  # Get image dimensions

    for i, box in enumerate(boxes):
        # Scale bounding box coordinates to image dimensions
        x_min, y_min, x_max, y_max = map(int, box)
        x_min = int(x_min * image_width)
        y_min = int(y_min * image_height)
        x_max = int(x_max * image_width)
        y_max = int(y_max * image_height)

        # Ensure bounding box coordinates are within image boundaries
        x_min = max(0, x_min)
        y_min = max(0, y_min)
        x_max = min(image_width, x_max)
        y_max = min(image_height, y_max)

        roi = image[y_min:y_max, x_min:x_max]

        # Check if ROI is empty before saving
        if roi.size != 0:
            roi_path = os.path.join(output_dir, label, f"{image_name}_roi{i}.jpg")
            cv2.imwrite(roi_path, roi)
        else:
            print(f"Warning: Empty ROI detected for box {i} in image {image_name}. Skipping.")



# Example usage for extracting ROIs
def process_image_for_vit(image_path, annotation_path, output_dir):
    # Read YOLO annotation file
    with open(annotation_path, 'r') as file:
        lines = file.readlines()

    boxes = []
    for line in lines:
        parts = line.strip().split()
        # YOLO format: <class_id> <x_center> <y_center> <width> <height>
        x_center, y_center, width, height = map(float, parts[1:])
        boxes.append([x_center - width / 2, y_center - height / 2, x_center + width / 2, y_center + height / 2])

    # Extract ROIs from the full image based on the bounding boxes
    extract_rois(image_path, boxes, output_dir, label="benign")  # Change "benign" as needed

# Process an image and extract ROIs
image_path = "/content/dataset/train/images/152_JPG.rf.0251cb7d865884a5ba118ad02192cefb.jpg"
annotation_path = "/content/dataset/train/labels/152_JPG.rf.0251cb7d865884a5ba118ad02192cefb.txt"
output_dir = "/content/dataset/vit_labels/train"
process_image_for_vit(image_path, annotation_path, output_dir)
```

```python
import torch
import torch.nn as nn
import torchvision.transforms as transforms
import torchvision.ops as ops

class YOLOViTModel(nn.Module):
    def __init__(self, yolo_model, vit_model, input_size=224):
        super(YOLOViTModel, self).__init__()
        self.yolo = yolo_model      # YOLOv4 for detection
        self.vit = vit_model        # Vision Transformer for classification
        self.input_size = input_size  # Size to resize ROIs for ViT (e.g., 224x224)

        # Transformation for ViT input
        self.resize_transform = transforms.Compose([
            transforms.Resize((input_size, input_size)),
            transforms.ToTensor()
        ])

    def forward(self, x):
        # Step 1: YOLOv4 detects objects and provides bounding boxes
        yolo_outputs, bboxes = self.yolo(x)

        # Store features of each ROI for classification by ViT
        roi_features = []
        for bbox in bboxes:
            # Step 2: Crop and resize each bounding box to feed into ViT
            roi = self.roi_pool(yolo_outputs, bbox)
            roi_features.append(self.vit(roi))

        # Step 3: Combine YOLO and ViT outputs
        return yolo_outputs, roi_features

    def roi_pool(self, features, bbox):
        """
        Implement ROI pooling or cropping for fixed-size ViT input.
        bbox should be in (x_min, y_min, x_max, y_max) format.
        """
        # Extract ROI from bounding box coordinates and resize to input size
        x_min, y_min, x_max, y_max = bbox
        cropped_feature = features[:, :, y_min:y_max, x_min:x_max]  # Crop ROI
        resized_feature = transforms.functional.resize(cropped_feature, (self.input_size, self.input_size))
        return resized_feature

# Define training function with loss functions for both detection and classification
def training_step(model, images, targets, detection_loss_fn, classification_loss_fn, alpha=1.0):
    """
    Perform a training step with combined detection and classification losses.
    :param model: YOLOViTModel instance
    :param images: Batch of images for training
    :param targets: Ground truth for detection and classification
    :param detection_loss_fn: Loss function for YOLO detection
    :param classification_loss_fn: Loss function for ViT classification
    :param alpha: Weight for the classification loss
    """
    # Forward pass
    yolo_outputs, vit_outputs = model(images)

    # Calculate detection loss
    detection_loss = detection_loss_fn(yolo_outputs, targets['detection'])

    # Calculate classification loss for each detected ROI
    classification_loss = sum([classification_loss_fn(output, target)
                               for output, target in zip(vit_outputs, targets['classification'])])

    # Combined loss
    total_loss = detection_loss + alpha * classification_loss
    return total_loss

# Example usage:
# Initialize the model with pretrained YOLO and ViT models
# yolo_model = <your pretrained YOLOv4 model>
# vit_model = <your pretrained ViT model>
# model = YOLOViTModel(yolo_model, vit_model)

# Example loss functions (adapt based on your models and data)
# detection_loss_fn = <your detection loss function>
# classification_loss_fn = nn.CrossEntropyLoss()  # Binary classification (benign/malignant)
# alpha = 1.0

# # Forward pass and training step
# images = <your batch of images>
# targets = {'detection': <detection targets>, 'classification': <classification labels>}
# loss = training_step(model, images, targets, detection_loss_fn, classification_loss_fn, alpha=alpha)
```

```python
import torch
from PIL import Image
import cv2
import numpy as np
from transformers import ViTForImageClassification, ViTFeatureExtractor

# Step 1: YOLO Model Setup (Using YOLOv8 from Ultralytics)
from ultralytics import YOLO

# Load YOLOv8 model (pre-trained or fine-tuned on your dataset)
yolo_model = YOLO("yolov8n.pt")  # Replace with your model path

# Step 2: ViT Model Setup
# Load pre-trained ViT model and feature extractor
vit_model = ViTForImageClassification.from_pretrained("google/vit-base-patch16-224")
vit_feature_extractor = ViTFeatureExtractor.from_pretrained("google/vit-base-patch16-224")

# Step 3: Load and Process the Input Image
def preprocess_image(image_path):
    """Load and preprocess the input image for YOLO."""
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB
    return image, image_rgb

# Step 4: Detect Tumors using YOLO
def detect_tumors(image):
    """Run YOLO detection and return bounding boxes."""
    results = yolo_model(image)
    boxes = results[0].boxes.xyxy.cpu().numpy()  # Extract bounding boxes
    scores = results[0].boxes.conf.cpu().numpy()  # Extract confidence scores
    return boxes, scores

# Step 5: Extract and Resize ROIs for ViT
def extract_and_resize_rois(image, boxes, input_size=224):
    """Extract ROIs and resize for ViT."""
    rois = []
    for box in boxes:
        x_min, y_min, x_max, y_max = map(int, box)
        roi = image[y_min:y_max, x_min:x_max]  # Crop ROI
        roi_resized = cv2.resize(roi, (input_size, input_size))  # Resize to ViT input size
        rois.append(roi_resized)
    return rois

# Step 6: Classify ROIs using ViT
def classify_rois(rois):
    """Classify each ROI using ViT."""
    results = []
    for roi in rois:
        # Convert ROI to PIL Image and normalize using ViT feature extractor
        roi_pil = Image.fromarray(roi)
        inputs = vit_feature_extractor(images=roi_pil, return_tensors="pt")
        outputs = vit_model(**inputs)
        predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
        predicted_class = torch.argmax(predictions, dim=-1).item()
        confidence = predictions[0][predicted_class].item()
        results.append((predicted_class, confidence))
    return results

# Step 7: Full Pipeline
def tumor_detection_and_classification(image_path):
    """Full pipeline: Detect tumors and classify them."""
    # Preprocess and detect tumors
    original_image, processed_image = preprocess_image(image_path)
    boxes, scores = detect_tumors(processed_image)

    # Extract and classify ROIs
    rois = extract_and_resize_rois(original_image, boxes)
    classifications = classify_rois(rois)

    # Display results
    for i, (box, score, classification) in enumerate(zip(boxes, scores, classifications)):
        x_min, y_min, x_max, y_max = map(int, box)
        label, confidence = classification
        print(f"Tumor {i + 1}:")
        print(f" - Bounding Box: [{x_min}, {y_min}, {x_max}, {y_max}]")
        print(f" - YOLO Confidence: {score:.2f}")
        print(f" - Classification: {label} (Confidence: {confidence:.2f})")

# Run the pipeline on an example image
image_path = "path_to_your_ultrasound_image.jpg"  # Replace with your image path
tumor_detection_and_classification(image_path)
```

```python
def classify_with_vit(img_path):
    # Read and convert the image with OpenCV
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Detect objects in the image
    detections = detect_objects(img_path)

    classifications = []
    for bbox in detections.xyxy:
        x1, y1, x2, y2 = map(int, bbox[:4])  # Get bounding box coordinates

        # Crop the detected region
        cropped_img = img_rgb[y1:y2, x1:x2]

        # Preprocess the cropped image for ViT
        vit_inputs = preprocess_vit_image(cropped_img)

        # Classify with ViT
        with torch.no_grad():
            vit_outputs = vit_model(**vit_inputs)
            predicted_class = vit_outputs.logits.argmax(-1).item()
            classifications.append(predicted_class)

    return classifications
```
